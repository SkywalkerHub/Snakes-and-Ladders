```python
import random #importing necessary functions
import sys
```

```python
#POSITION OF SNAKES
snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}

#POSITION OF LADDERS
ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}
```

```python
#GETTING THE NAME OF THE PLAYERS
def get_player_name():
    while True:
        player1 = input("Enter the name of player 1: ")
        #Checking if name is empty or only whitesapces
        if(not(player1 and player1.strip())):
            print("Invalid input: Name cannot be empty or whitespace.")
        else:
            break
    while True:
        player2 = input("Enter the name of player 2: ")
        #Checking if name is empty or only whitesapces
        if(not(player2 and player2.strip())):
            print("Invalid input: Name cannot be empty or whitespace.")
        #Checking for dupliate names
        elif (player1.strip() == player2.strip()):
            print("Cannot have duplicate names.")
        else:
            break
    return player1, player2
```

```python
# Simulate a dice roll with random function
def dice_roll():
    dice_value = random.randint(1,6)
    print(f"You got: {dice_value}")
    return dice_value
```

```python
# Getting bit by a snake
def snake_bite(old_pos, current_pos, player_name):
    print(f"{player_name} got bit by a snake and moved down from {old_pos} to {current_pos}")

# Climbing a ladder
def ladder_climb(old_pos, current_pos, player_name):
    print(f"{player_name} took a ladder and climbed up from {old_pos} to {current_pos}")
```

```python
def player_moves(player_name, current_pos, dice_value):
    old_pos = current_pos
    current_pos = int(current_pos) + int(dice_value)
    # overshoot condition- If final position becomes greater than 100.
    if current_pos > 100:
        print(f"Can't move! You need {100-old_pos} to win the game.")
        return old_pos
    print(f"{player_name} moved from {old_pos} to {current_pos}")
    # Getting bit by a snake condition
    if current_pos in snakes:
        final_pos = snakes.get(current_pos)
        snake_bite(current_pos, final_pos, player_name)
    #Getting a ladder condition
    elif current_pos in ladders:
        final_pos = ladders.get(current_pos)
        ladder_climb(current_pos, final_pos, player_name)
    else:
        final_pos = current_pos
    return final_pos
```

```python
def check_win(player_name, position):
    if position == 100:
        print(f"{player_name} WON THE GAME !!!")
        print("GAME SUCCESSFULLY FINISHED \n MADE BY SHUBHAM :)")
        sys.exit()
```

```python
# START OF THE GAME
print("WELCOME TO SNAKE AND LADDERS GAME")
player1, player2 = get_player_name()

player1_current_pos = 0
player2_current_pos = 0

while True: 
    #selecting the game mode from the player
    while True:
        player1_input = input(f"{player1} type 'Roll' to roll a die \n'Manual' for Manual-roll \n'quit' to forfeit the match \n")
        if player1_input == "Roll":                                  #DICE-ROLL BASED GAME MODE
            dice_value = dice_roll()
            player1_current_pos = player_moves(player1, player1_current_pos, dice_value)
            print(f"Your final position is {player1_current_pos}")
            check_win(player1, player1_current_pos)
            break 
        elif player1_input == "Manual":                                #MANUAL GAME MODE
            dice_value = input("Enter a number between 1 and 20.")
            if int(dice_value) < 1 or int(dice_value) > 20:
                print("Invalid input")
                
            else:                                                      
                print(f"You chose {dice_value}")
                player1_current_pos = player_moves(player1, player1_current_pos, dice_value)
                print(f"Your final position is {player1_current_pos}")
                check_win(player1, player1_current_pos)
                break
        elif player1_input == "quit":                                  #quit condition
            print(f"{player1} quits.\n {player2} WON THE GAME !!!")
            sys.exit()
        else:
            print("Invalid input")
   
    while True:
        player2_input = input(f"{player2} type 'Roll' to roll a die \n'Manual' for Manual-roll \n'quit' to forfeit the match \n")
        if player2_input == "Roll":
            dice_value = dice_roll()
            player2_current_pos = player_moves(player2, player2_current_pos, dice_value)
            print(f"Your final position is {player2_current_pos}")
            check_win(player2, player2_current_pos)
            break
        elif player2_input == "Manual":
            dice_value = input("Enter a number between 1 and 20.")
            if int(dice_value) < 1 or int(dice_value) > 20:
                print("Invalid input")
                
            else:
                print(f"You chose {dice_value}")
                player2_current_pos = player_moves(player2, player2_current_pos, dice_value)
                print(f"Your final position is {player2_current_pos}")
                check_win(player2, player2_current_pos)
                break
        elif player2_input == "quit":
            print(f"{player2} quits.\n {player1} WON THE GAME !!!")
            sys.exit()
        else:
            print("Invalid input")
```
